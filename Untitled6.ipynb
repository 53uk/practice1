```python
import easyocr
import numpy as np
import cv2
import random
import requests
from PIL import ImageFont, ImageDraw, Image
import matplotlib.pyplot as plt

# EasyOCR Reader 생성
reader = easyocr.Reader(['ko', 'en'], gpu=True)

# 이미지 URL
img_url = 'https://www.dropbox.com/scl/fi/qn18l5ovjf6xito31cl4x/KakaoTalk_20240329_153742049.jpg?rlkey=5pwde2ic4q318ejp5ye8tky6r&st=c37muy8a&raw=1'
img_response = requests.get(img_url).content
img_array = np.array(bytearray(img_response), dtype=np.uint8)
img = cv2.imdecode(img_array, -1)

# EasyOCR로 텍스트 인식
result = reader.readtext(img)

# PIL 이미지로 변환
img = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
font = ImageFont.truetype('NanumGothic-Regular.ttf', 50)
draw = ImageDraw.Draw(img)

# 랜덤 컬러 생성
np.random.seed(42)
COLORS = np.random.randint(0, 255, size=(100, 3), dtype="uint8")

# 검출된 텍스트 위에 사각형 및 텍스트 그리기
for i in result:
    x = i[0][0][0]
    y = i[0][0][1]
    w = i[0][1][0] - i[0][0][0]
    h = i[0][2][1] - i[0][1][1]

    color_idx = random.randint(0, 99)  # 0~99 사이에서 컬러 인덱스 랜덤 선택
    color = [int(c) for c in COLORS[color_idx]]

    draw.rectangle(((x, y), (x + w, y + h)), outline=tuple(color), width=2)
    draw.text((int((x + x + w) / 2), y - 2), str(i[1]), font=font, fill=tuple(color))

# 결과 이미지 표시
plt.figure(figsize=(12, 12))
plt.imshow(img)
plt.axis('off')  # 축 표시 제거
plt.show()

```
